```python
# console version
# see line 48 to set the language you wanna play in

from unidecode import unidecode
from random import choice
from time import sleep

class colors:
    '''Colors class:reset all colors with colors.reset; two
    sub classes fg for foreground
    and bg for background; use as colors.subclass.colorname.
    i.e. colors.fg.red or colors.bg.greenalso, the generic bold, disable,
    underline, reverse, strike through,
    and invisible work with the main class i.e. colors.bold'''
    reset='\033[0m'
    bold='\033[01m'
    disable='\033[02m'
    underline='\033[04m'
    reverse='\033[07m'
    strikethrough='\033[09m'
    invisible='\033[08m'
    class fg:
        black='\033[30m'
        red='\033[31m'
        green='\033[32m'
        orange='\033[33m'
        blue='\033[34m'
        purple='\033[35m'
        cyan='\033[36m'
        lightgrey='\033[37m'
        darkgrey='\033[90m'
        lightred='\033[91m'
        lightgreen='\033[92m'
        yellow='\033[93m'
        lightblue='\033[94m'
        pink='\033[95m'
        lightcyan='\033[96m'
    class bg:
        black='\033[40m'
        red='\033[41m'
        green='\033[42m'
        orange='\033[43m'
        blue='\033[44m'
        purple='\033[45m'
        cyan='\033[46m'
        lightgrey='\033[47m'

language = 'ptbr' # ptbr or en

with open('words_' + language +'.txt', 'r', encoding='utf-8') as a:
    words = [unidecode(x.upper()) for x in a.read().split(", ")]
word = choice(words)
with open('accepted_words_' + language +'.txt', 'r', encoding='utf-8') as a:
    accepted_answers = [unidecode(x.upper()) for x in a.read().split(", ")]
words.extend(accepted_answers)

invalids = []
attempts = 0
while True:
    if attempts > 5:
        print(f'game over, word was {word}')
        break
    while True:
        guess = unidecode(str(input('guess: ')).upper())
        if len(guess) == 5:
            break
        else:
            print('favor inserir palavra de 5 letras')
    if guess == word:
        for i, v in enumerate(guess):
            if v == word[i]: #letra na posicao certa
    #             print(f'{v} is in correct position')
                print(colors.bg.green, v, colors.reset, end='')
        print('\nyou won')
        break
    else:
        for i, v in enumerate(guess):
            if v == word[i]: #letra na posicao certa
    #             print(f'{v} is in correct position')
                print(colors.bg.green, v, colors.reset, end='')
            elif v in word and word.count(v) >= guess.count(v): #esta na palavra
    #             print(f'{v} is in word')
                print(colors.bg.orange, colors.fg.black, v, colors.reset, end='')
            elif v in word and word.count(v) < guess.count(v) and i < guess.rfind(v):
                print(colors.bg.orange, colors.fg.black, v, colors.reset, end='')
            else:
                if v not in invalids:
                    invalids.append(v)
    #             print(f'{v} is not in word')
                print(colors.bg.black, colors.fg.lightgrey, v, colors.reset, end='')
        attempts += 1
    print()
    print(f'letters not on word: {invalids}')
    print()
```
